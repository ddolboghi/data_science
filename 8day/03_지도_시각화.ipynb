```python
# 지도 그리기
starbucks_map = folium.Map(
    # 최초에 보여줄 지도 위치 지정(지도 중심점 좌표)
    # [위도, 경도] 지정
    location=[37.573050, 126.979189],
    # 지도 스타일 지정
    # tiles: 도시형 건물 등 일반적으로 사용되는 스타일
    # - OpenStreetMap: 도시형 건물 등 일반적으로 사용되는 스타일
    # - Stamen Terrain: 산림 위주로 보여주는 스타일
    # - cartodbpositron: 하천이나 도로(길) 위주로 보여주는 스타일
    tiles="OpenStreetMap",
    # 최초 보여질 지도 zoom 사이즈 지정
    zoom_start=50 
)
starbucks_map
```

```python
"""
스타벅스 매장 위치 표시
- 지도 시각화 시 여러 좌표의 위치를 표시할때는 반복문 사용
"""
for idx in seoul_starbucks.index:
    if seoul_starbucks.loc[idx, "시군구명"] == "강남구":
        lat = seoul_starbucks.loc[idx, "위도"]
        lng = seoul_starbucks.loc[idx, "경도"]

        #지도에 표시할 마킹 모양 등 스타일 지정
        folium.CircleMarker(
            # 마킹 위치 지정
            location = [lat, lng],
            fill_color="green",
            fill_opacity=1, # 불투명도
            color="yellow",
            weight=1,
            radius=3
        ).add_to(starbucks_map)# 사용할 지도형태 지정
```

```python
"""
매장 타입별로 위치 표시
"""
starbucks_map2 = folium.Map(
    location=[37.573050, 126.979189],
    tiles="OpenStreetMap",
    zoom_start=50 
)
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx, "위도"]
    lng = seoul_starbucks.loc[idx, "경도"]
    store_type = seoul_starbucks.loc[idx, "매장타입"]
    
    # 매장 타입별 표시할 색상 지정
    fillColor=""
    size=1
    if store_type == "general":
        fillColor = "gray"
        size = 3
    elif store_type == "reserve":
        fillColor = "blue"
        size = 5
    elif store_type == "generalDT":
        fillColor = "red"
        size = 5
        
    #표시할 마커 스타일 지정
    folium.CircleMarker(
        location = [lat, lng],
        fill_color=fillColor,
        fill=True, # 마커를 무조건 채움
        fill_opacity=1,
        color=fillColor,
        weight=1,
        radius=size
    ).add_to(starbucks_map2)

starbucks_map2
```
