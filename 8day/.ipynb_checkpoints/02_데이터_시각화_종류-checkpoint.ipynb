# subplot
- fig = plt.figure()
- ax = figure객체.add_subplot(행, 열, 위치값)
- figure객체.tight_layout(): subplot간 간격 조정
- figure객체.suptitle("제목"): 전체 그래프의 제목


- 우상향 데이터: 왼쪽 밑에서 오른쪽 위로 분포
- 우하향 데이터: 왼쪽 위에서 오른쪽 아래로 분포
### --> 여러 컬럼을 비교할때 컬럼들간 상관관계 확인

```python
# data1~data4 그래프를 4개의 공간에 동시에 그리기
fig = plt.figure()

# 각각의 그래프를 그리기 위한 공간 만들기
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# subplot에 제목 넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")

# 데이터를 넣어서 그림 그리기: 점으로 표시
ax1.plot(data1["x"], data1["y"], "o", c="r")
ax2.plot(data2["x"], data2["y"], "o", c="g")
ax3.plot(data3["x"], data3["y"], "o", c="b")
ax4.plot(data4["x"], data4["y"], "o", c="y")

fig.suptitle("Anscombe Data")
fig.tight_layout()
```

```python
plt.title("Legend")
plt.plot([1, 2, 3, 4], label="asc", c="red")# 증가
plt.plot([4, 3, 2, 1], label="desc", color="green")# 감소
# label에 대한 범례 추가
plt.legend()
# x축, y축 이름 지정
plt.xlabel("index")
plt.ylabel("data")
plt.show()
```

```python
# 선 모양 변경하기
plt.plot([1, 2, 3, 4], label="line", linestyle="-", c="r", marker="d")
plt.plot([4, 3, 2, 1], label="dash", linestyle="--", c="#1344", marker="x")
plt.plot([2, 1, 1, 2], label="dot", linestyle=":", c="magenta", marker="o")
plt.legend()
plt.show()
```

```python
"""
c: 색상
lw: 선 굵기
ls: 선 스타일
marker: 값의 위치에 표시할 마커 형태(o, s, x)
ms: 마커 크기
mfc: 마커 내부 채우기 색
mec: 마커 테두리 색
mew: 마커 데투리 크기
"""
plt.title("x, y 레인지")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 
         c="b", lw=1, ls="--", marker="o", ms=10, mec="g", mew=3, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

# 데이터 읽어들인후 시각화

```python
tips = sns.load_dataset("tips")
tips.info()
tips.describe()
```

## 데이터프레임에서 시각화 라이브러리 사용

```python
# DataFrame에서 제공해주는 기본 [선그래프]

# 선그래프는 kind 생략 가능, dafault: 탐색적 데이터 분석(EDA)시 사용
tips.plot(kind="line")

# 히스토그램: 현황 분석시 사용
tips.plot(kind="hist")

# 박스플롯: 이상치 분석시 사용
tips.plot(kind="box")
```

## matplotlib 사용

```python
# 선그래프
plt.figure(figsize=(3, 3))
plt.plot(tips.total_bill)
```

```python
# 막대그래프
plt.figure(figsize=(3, 3))
plt.bar(tips.sex, tips.total_bill)
```

```python
"""
산점도 그래프
- 점들이 산만하게 흩어진 그래프
- 두 데이터 간의 관계 확인
- 데이터의 분포 확인 가능, 이상치 데이터를 대략 확인 가능
"""
plt.scatter(tips.total_bill, tips.tip)
```

```python
"""
히스토그램
- 범주형 데이터 빈도분석시 주로 사용
"""
plt.hist(tips.total_bill)
```

```python
# stripplot 모양 지정
sns.stripplot(x="day", y="total_bill", data=tips)
# 제목 넣기
plt.title("팁 데이터")
plt.xlabel("요일")
plt.ylabel("요금")
plt.show()
```

```python
"""
막대그래프
- 평균치를 빠르게 집계해서 시각화
- 막대그래프 위의 선: 오차막대(error bar)
- 오차막대는 반만 막대그래프에 걸쳐있음
- 표준편차가 클수록 오차막대 길이가 길어짐
- 오차막대 길이가 작을수록 좋음
"""
sns.barplot(x="sex", y="tip", data=tips)
```

```python
# 오차막대 숨기기: estimator=len
sns.barplot(x="sex", y="tip", estimator=len,data=tips)
```

```python
"""
3개 컬럼의 데이터 이용
- 성별/요일별 tip데이터 시각화
- hue= 범주화시킬 데이터
"""
sns.barplot(x="sex", y="tip", hue="day", data=tips)
```

```python
"""
박스플롯 시각화
- orient="v": vertical(수직)으로 표현, defualt
- orient="h": horizen(수평)으로 표현
"""
sns.boxplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(data=tips, orient="h")
```

```python
"""
커널 밀도 그래프
- 히스토그램보다 부드러운 형태의 분포 곡선
- (데이터.컬럼)을 인자로 넘겨줘야함
"""
sns.kdeplot(tips.tip)
```

```python
"""
커널밀도그래프와 히스토그램을 하나의 그래프로 표현하는 함수
사라질 예정
- 커널밀도그래프를 벗어난 히스토그램은 이상치로 볼수 있음

히스토그램은 막대그래프와 달리 각 항목이 붙어있음!
"""
sns.histplot(tips.tip)
```

```python
"""
countplot: 범주형 데이터 표시
- data=DataFrame
- 특정 컬럼의 범주 데이터 별로 빈도 확인 가능
"""
sns.countplot(x="day", data=tips)
```

```python
"""
여러 그래프를 하나로 그리기
- kind=그래프종류 지정
  - scatter:산점도와 각 축 기준 산점도에 따른 막대 그래프를 동시에 시각화
  - kde: 커널밀도와 선그래프를 동시에 시각화, 등고선 간격이 좁을수록 밀집도 높아짐
"""
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

```python
"""
다차원 복합 데이터 시각화
catplot: 색상과 행 등을 동시에 사용하여 3개 이상의 범주값의 분포 변화를 시각화
- kind=
  - swarm: 데이터를 나타내는 점이 겹치지 않도록 표현
  - violin: swarm의 점의 전체 모양을 바이올린 모양으로 표현
"""
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
"""
다차원 시각화: 여러 컬럼의 데이터 시각화
- 데이터프레임 자체를 넘겨줌
- 크로스 체크 가능
- 상관관계 분석시 주로 사용 --> 그래프가 이쁘면 관계성이 높음
- 막대그래프 = 컬럼 자기자신
"""
sns.pairplot(tips)
```

```python
"""
다차원 상관도
- hue=범주형 컬럼 --> 범주의 분포 확인
"""
sns.pairplot(tips, hue="sex")
```
