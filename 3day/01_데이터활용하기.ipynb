## pd.merge()
: 컬럼 단위로 통합
- left:  통합할 기준 데이터 프레임(이하 df)
- right: 가져올 df
- how: 어떻게 가져올지 지정
  - inner: 같은 것만
  - left: 왼쪽 df 전체가져옴, 오른쪽 df와 다른 값은 NaN --> 값 손실 방지 위해 주로 쓰임
- left_on: 왼쪽 df에서 오른쪽 df과 비교할 컬럼 지정
- right_on: 오른쪽 df에서 왼쪽 df와 비교할 컬럼 지정

```python
sample_1_code = pd.merge(left = sample_1,
                        right = code_master,
                        how = "left",
                        left_on = "국적코드",
                        right_on = "국적코드")
sample_1_code
```

# [문제]files_all 의 모든 데이터 통합하기
- 국적코드 국적명으로 바꾸기
- 최종 결과 변수명 : sample_all
- 최종 결과 파일명 : sample_all.xlsx
- 파일 저장위치 : files_01_new 폴더에...
- 기준년도 : 2019-01 부터 시작합니다.

```python
def get_sample(i, k_plus):
    mm = i % 12 + 1
    yyyy = 2018 + k_plus
    
    file_path = "./files_01_all/sample({}).xlsx".format(i)
    sample = pd.read_excel(file_path, header = 1, skipfooter = 2, usecols = "A:C")
    
    baseYM = "{}-{}".format(str(yyyy), str(mm).zfill(2))
    sample["기준년월"] = baseYM
#     print(yyyy, mm)

    # code_master 호출
    code_master = pd.read_excel("./files_01/sample_codemaster.xlsx",)

    # sample과 code_master 병합
    sample_code = pd.merge(left = sample,
                                    right = code_master,
                                    how = "left",
                                    left_on = "국적코드",
                                    right_on = "국적코드")

    return sample_code
```

```python
import pandas as pd
sample_all = pd.DataFrame()
k_plus = 0
n = 0
def sample_final(n):
    global sample_all
    global k_plus
    if n == 31:
        return sample_all
    if (n % 12 + 1) == 1:
        k_plus += 1
    sample_return = get_sample(n, k_plus)
    sample_all = pd.concat([sample_all, sample_return], ignore_index=True)
    n += 1
    sample_final(n)

sample_final(n)
sample_all
```

```python
sample_all.to_excel("./files_01_all/sample_all.xlsx", index=False)
```

## 답안

```python
### 함수 만들기
def getNewDataFrame(p_i, p_yyyy, p_mm) :
    file_path = "./files_01_all/sample({}).xlsx".format(p_i)
    sample_temp = pd.read_excel(file_path, header=1, skipfooter=2, usecols="A:C")

    ### 기준년월 생성
    baseYM = str(p_yyyy) + "-" + str(p_mm).zfill(2)
    sample_temp["기준년월"] = baseYM

    code_master = pd.read_excel("./files_01/sample_codemaster.xlsx",)
    # 국적명 가지고 오기
    sample_end = pd.merge(left=sample_temp, 
                           right=code_master,
                           how='left',
                           left_on='국적코드',
                           right_on='국적코드')
    return sample_end
```

```python
### 함수 호출
sampleAll = pd.DataFrame()

i = 0;
for yyyy in range(2019, 2022, 1) :
    for mm in range(1, 13, 1) :
        try :
            sample_return = getNewDataFrame(i, yyyy, mm)
            sampleAll = pd.concat([sampleAll, sample_return], ignore_index=True)
            i += 1
        except :
            pass

print(len(sampleAll))
sampleAll
```
