# 전처리
: 결측치 데이터, 중복 데이터, 이상치 데이터 처리
- 결측치: 컬럼에 값이 없음
- 중복: 행들의 값들이 중복된 경우
- 이상치: 특정 범위를 벗어나는 값들

--> 일반적인 처리: 삭제, 값 채우기 <br>
--> 이상치 처리: 이상치 산출 계산 공식 사용해서 처리

## csv unicodeerror
- 데이터가 utf-8 형식이 아닐 경우 발생
- read_csv(..., encoding='euc-kr' or 'cp949')
- cp949는 euc-kr의 확장 버전

```python
import pandas as pd
import numpy as np
file_path = "./data/bicycle.csv"
df = pd.read_csv(file_path, encoding='cp949')
df
```

```python
df.info()
```

# isnull()
: 결측치가 True로 나옴

```python
df.isnull()
```

# notnull()
: 결측치가 아닌 것이 True로 나옴

```python
df.notnull()
```

```python
# 각 컬럼별 결측치 개수 확인
# sum(0) --> 행 단위
df.isnull().sum(0)
```

```python
# 행 단위 결측데이터 컬럼 개수 확인
# sum(1) --> 컬럼의 개수
df.isnull().sum(1)
```

# 결측치 삭제
## .dropna(axis=)
- 행 단위 삭제: axis=0
- 컬럼 단위 삭제: axis=1

```python
df_drop_allrow = df.dropna(axis=0)
df_drop_allrow.info()
```

```python
df_drop_allcol = df.dropna(axis=1)
df_drop_allcol.info()
```

## 특정 컬럼 기준 삭제

```python
df["대여소명"].dropna(axis=0)
```

```python
df[["대여거치대"]].dropna(axis=0)
```

```python
df[["대여소번호", "대여거치대", "이용시간"]].dropna(axis=0)
```

# 결측치 채우기


## .fillna(0)
: 결측치를 0으로 채움

```python
df.fillna(0).info()
```

```python
df["이용거리"].fillna("missing")
```

```python
df["이용거리"].fillna(df["이용거리"].mean())
```
