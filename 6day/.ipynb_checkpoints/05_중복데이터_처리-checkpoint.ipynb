```python
import pandas as pd
import numpy as np
file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path, encoding="euc-kr")
df.info()
```

# duplicated() 
- 컬럼들이 다음 행에서 중복이 있는지 확인
- return: bool

```python
# 전체 컬럼에서 다음 행이 중복되는지 확인
df[df.duplicated()]
```

```python
df["이용거리"]
```

```python
# 특정 컬럼에서 중복되는 행 추출 
df[df[["이용거리", "나이"]].duplicated()]
```

```python
# df[["이용거리", "나이"]]
df[df.duplicated(["이용거리", "나이"])]
```

# 중복데이처 처리 방법 지정
## keep속성 사용
- first: 첫번째 중복데이터 살리고(False), 아래 나머지(True)
- last: 마지막 중복데이터는 살리고(False), 위에 나머지(True)
- 생략가능: default first
- 가장 많이 사용하는 방식

```python
df.duplicated(["이용거리"], keep="last")
```

# drop_duplicates()
: 중복데이터 삭제

```python
df.drop_duplicates(["이용거리"], keep="first")
```

```python
# 중복이 있는 모든 데이터 삭제
df.drop_duplicates(["이용거리"], keep=False)
```
