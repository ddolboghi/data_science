### merge()
```
# 기준열 이름이 같을 때
pd.merge(left, right, on = '기준열', how = '조인방식')

# 기준열 이름이 다를 때
pd.merge(left, right, left_on = '왼쪽 열', right_on = '오른쪽 열', how = '조인방식')
```
- inner: 두 df에서 같은 조건에 대해서만 join--> 알아서 같은 컬럼에 대해 수행, A,B교집합
- outer: 왼쪽 df 기준으로 모든 데이터 가져옴, 오른쪽 df 기준으로 모든 데이터 가져옴, A,B합집합 
- left: 왼쪽 df 전체, A집합
- right: 오른쪽 df 전체, B집합 

````python
df_info1 = pd.DataFrame({"고객번호":[1001, 1002, 1003, 1004, 1005, 1006, 1007], 
                       "이름": ["강", "김", "홍", "이", "박", "최", "소"]}, 
                       columns=["고객번호", "이름"])
df_info1
````

````python
df_info2 = pd.DataFrame({"고객번호":[1001, 1001, 1005, 1006, 1008, 1001], 
                       "금액": [10000, 20000, 15000, 5000, 100000, 300000]}, 
                       columns=["고객번호", "금액"])
df_info2
````

````python
pd.merge(df_info1, df_info2, how="inner")
````

````python
pd.merge(df_info1, df_info2, how="outer")
````

````python
pd.merge(df_info1, df_info2, how="left")
````

````python
pd.merge(df_info1, df_info2, how="right")
````
