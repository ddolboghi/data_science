```python
import pandas as pd
file_path = "./data/exam_sample.csv"
df = pd.read_csv(file_path, encoding="cp949")
df.info()
df.describe()
```

# .groupby(["컬럼명"])
- 그룹화할 컬럼이 1개면 대괄호 안써도 됨
- 그룹화한 데이터는 head()로 볼수 있음
- 각 그룹별로 따로따로 인덱스 재지정됨
- .head(n): 각 그룹별 n번째 값까지 추출
- .mean(): 그룹 지어 표 만들때 유용함

```python
df1 = df.groupby(["class"])
df1.head()
```

## 그룹화된 데이터프레임.get_group("그룹 이름")
- 그룹 이름에 해당하는 데이터만 추출

```python
df1.get_group("A")
```

```python
# 특정 컬럼을 기준으로 그룹화 후 그룹별 평균 구하기
df.groupby("class").mean()
```

```python
# 성별을 그룹화
df_temp = df.groupby(["class", "sex"])
df_temp.head(1)
```

```python
df.groupby(["class", "sex"]).mean()
```

## 그룹화할때 여러가지 컬럼 사용 위치
- class별 math 평균 
- 그룹에서 추출할 컬럼명을 2차원으로 지정하면 데이터프레임 형태로 추출됨

```python
# 방법1 --> 가장 많이 사용
df.groupby("class")["math"].mean()
```

```python
# 방법2
df[["math"]].groupby(df["class"]).mean()
```

```python
# 방법3
df.groupby(["class"]).mean()["math"]
```

```python
# 반별 수학 응시생 수
df.groupby("class")[["math"]].count()
```

```python
# 데이터프레임 형태로 보여주기
df.groupby(["class"]).count()[["math"]]
```

```python
df[["class"]]
```

```python
# 성별 수학 점수의 평균을 데이터프레임 형태로 추출
df.groupby("sex")[["math"]].max()
```

```python
# 성별 그룹화
sex_group = df.groupby("sex")
sex_group.head()
```

# 그룹화데이터.groups
: {그룹이름:[그룹별 인덱스 값], ...}

```python
# 그룹 내부구조 확인하기
sex_group.groups
```

```python
# 위 groups의 형태를 이용해서 "m"에 있는 모든 인덱스 값 추출후 하나씩 출력
# arr = sex_group.groups["m"]
# for x in arr:
#     print(x)
    
for k in sex_group.groups.keys():
    if k != "m":
        continue
    for i in sex_group.groups[k]:
        print(i)
```

# 데이터 재구조화

## pd.cut()
: 특정 컬럼의 데이터를 특정 구간으로 나눔
- 동일한 길이(값의 범위)로 나눔
- labels=False: 구간별 인덱스 번호처럼 0부터 시작하는 번호 부여, 구간 값 직접 지정 가능

```python
df1 = pd.cut(df["math"], 3, labels=False)
df1
```

```python
# 수학점수를 구간별로 그룹화
df2 = df["math"].groupby(df1)
df2.head(1)
```

```python
df2.get_group(1)
```

# 그룹화한 데이터 기초통계
## agg(["count", "mean", "std", "min", "max"])


```python
df2.agg(["count", "mean", "std", "min", "max"])
```

# pd.qcut()
- 유사한 값들(동일한 범위가 아닌 유사한 값들)끼리 구간 나누기
- 구간이 많아지면, 해당 구간에 없는 값이 생길 수도 있음
- 해당 구간에 존재하는 값이 없을 경우에는 조회 안됨

```python
pd.qcut(df["math"], 5, labels=False)
```

```python
import numpy as np
pd.qcut(df["math"], 5, labels=np.arange(1, 6, 1))
```

# one-hot encoding
# pd.get_dummies(df)
: 범주를 컬럼으로 데이터화할때 사용, 범주로 표현될수 있는 모든 데이터들을 0과 1로 표현
- 컬럼으로 변환되면 시각화에 유용함

```python
df_onehot = pd.get_dummies(df)
df_onehot
```

```python
# 특정 컬럼에 대해 one-hot encoding
df_onehot2 = pd.get_dummies(df["sex"])
df_onehot2
```

```python
# df에서 반별 성별 과학점수의 평균 조회
# 반은 행, 성별은 컬럼
pd.pivot_table(df, values="science", index="class", columns="sex", aggfunc="mean")
```

```python
# 반별 성별로 그룹화하여 과학점수의 평균을 구하고 데이터프레임 형태로 조회
df.groupby(["class", "sex"])[["science"]].mean().reset_index()
```
