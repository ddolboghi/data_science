# 승하차 정류장별 체류시간(분) 상위 30건 분석 비교

```python
# 승하차 정류장 컬럼 추가
in_out_list = []
for i in range(0, len(df_sort)):
    in_out_list.append(df_sort["승차정류장"][i] + "-" + df_sort["하차정류장"][i])
print(len(in_out_list))
df_sort["승차-하차정류장"] = in_out_list
df_sort.head(3)
```

```python
# timedelta 안쓰는 방법

h_start = ((df_sort["승차시각"] // 10000)%100)
h_end = ((df_sort["하차시각"] // 10000)%100).replace(0, 24)

m_start = ((df_sort["승차시각"] // 100)%100)
m_end = ((df_sort["하차시각"] // 100)%100)

s_start = (df_sort["승차시각"] % 100)
s_end = (df_sort["하차시각"] % 100)

(h_end - h_start) * 3600
```

```python
from datetime import timedelta
tt1 = timedelta(hours=1, minutes=29, seconds = 59)
tt2 = timedelta(hours=1, minutes=1, seconds = 25)
type(tt1-tt2)
```

```python
"""
timedelta.total_seconds()로 체류시간을 초단위로 변환
"""
from datetime import timedelta
dif_list = []

for i in range(0, len(df_sort)):
    day_start = int(str(df_sort["승차시각"][i])[6:8])
    day_end = int(str(df_sort["하차시각"][i])[6:8])
    
    h_start = int(str(df_sort["승차시각"][i])[8:10])
    h_end = int(str(df_sort["하차시각"][i])[8:10])
    if day_start != day_end:
        h_end = 24

    m_start = int(str(df_sort["승차시각"][i])[10:12])
    m_end = int(str(df_sort["하차시각"][i])[10:12])

    s_start = int(str(df_sort["승차시각"][i])[12:14])
    s_end = int(str(df_sort["하차시각"][i])[12:14])

    end_time = timedelta(hours=h_end, minutes=m_end, seconds = s_end)
    start_time = timedelta(hours=h_start, minutes=m_start, seconds = s_start)

    dif_time = end_time - start_time
#     dif_str = str(end_time - start_time)
    dif_s = dif_time.total_seconds()
    dif_list.append(dif_s)
    
len(dif_list)
```

```python
df_sort["체류시간(s)"] = dif_list
```

```python
# 개인별 체류시간 top30
df_sort.sort_values(by="체류시간(s)", ascending=False).head(30)
```

```python
"""
'승차-하차정류장'컬럼의 값이 같은 것들은 그 평균으로 대체해야함
승차-하차정류장 기준으로 체류시간(s)의 평균 구하기
pivot_table(index="승차-하차정류장", values="체류시간(s)", aggfunc="mean")
--> '승차-하차정류장'과 '체류시간(s)'만 추출
groupby(["승차-하차정류장"]).mean()
--> '승차-하차정류장'기준으로 모든 열에 대한 연산
"""
# df_pivot2 = df_sort.pivot_table(index="승차-하차정류장", 
#                                 values="체류시간(s)", 
#                                 aggfunc="mean").reset_index(drop=False)
# df_pivot2
df_mean = df_sort.groupby(df_sort["승차-하차정류장"]).mean()
df_final = df_mean.reset_index(drop=False)
print(len(df_final))
df_final
```

```python
""" 
loc에 조건줘서 df_sort에서 '승차정류장==하차정류장'인 '체류시간'만 추출하기

추출한 데이터프레임에서 각 정류장마다 체류시간을 초단위로 변환후 평균 구하기
다시 체류시간 형식으로 변환 후 값 교체

약 3백만번 연산함
"""
condition1 = df_sort["승차정류장"] == df_sort["하차정류장"]
df_same = df_sort[condition1].reset_index()
df_same_origin = df_same["승차정류장"].unique()
ss_tot_all = 0
count = 0
print(len(df_sort), len(df_same), len(df_same_origin))
# df_same에서 df_same_unique에있는 값들로 루프돌면서 평균 구하고 변환
for origin in df_same_origin:
    condition2 = df_sort["승차정류장"] == origin
    
    for i in range(0, len(df_same)):
        busstop = df_same["승차정류장"][i]
        
        if busstop == origin:
            count += 1
            hh, mm, ss = df_same["체류시간"][i][-7:].split(":")
            ss_tot = int(hh) * 3600 + int(mm)* 60 + int(ss)
            ss_tot_all += ss_tot
#             print(ss_tot_all, count)
            
    s_mean = round(ss_tot_all / count)
    count = 0
    ss_tot_all = 0
    
    h_r = str(s_mean // 3600)
    m_div = s_mean // 60
    if m_div >= 60:
        m_r = str(m_div-60).zfill(2)
    else:
        m_r = str(m_div).zfill(2)
    s_r = str(s_mean % 60).zfill(2)

    t_r = h_r + ":" + m_r + ":" + s_r
    df_sort.loc[condition1 & condition2, "체류시간"] = t_r
df_sort[condition1]["체류시간"]
```
